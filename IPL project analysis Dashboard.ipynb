```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data =pd.read_csv("IPL 2022.csv")
```

```python
data
```

# Number of matches won by IPL 2022 Team

```python
figure = px.bar(data, x=data['match_winner'],title ="Number of matches win in IPl 2022")
figure.show()
```

```python
data['won_by'] =data['won_by'].map({'Wickets': 'Chasing',
                                   'Runs' : 'Defending'})
won_by = data["won_by"].value_counts()
label =won_by.index
counts = won_by.values
colors = ['red','lightgreen']
fig =go.Figure(data = [go.Pie(labels =label , values=counts)])
fig.update_layout(title_text = "Number of matches won by defending or chasing")
fig.update_traces(hoverinfo = 'label+percent' ,textinfo = 'value' , textfont_size=30 , marker=dict(colors = colors ,line =dict(color = 'black',width=3)))
```

```python
figure = px.bar(data, x=data["best_bowling"], title = "Best Bowler in IPL 2022")
figure.show()
```

```python
figure =px.bar(data ,x= ['player_of_the_match'],
               title ="Most plyer of match awards")
figure.show()
```

# top scorer in 2022 

```python
figure = px.bar(data, x=data["top_scorer"],
                y =data["highscore"],
                color= data["highscore"],title ="Top scorer in IPL 2022")
figure.show()
```
